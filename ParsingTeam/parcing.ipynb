```python
import json
import re
from parse_func import parse_func
```

```python
not_words = ["not", "no", "can't"]
```

```python
data = []
with open('AI2-8thGr-NDMC-Feb2016-Train.jsonl') as f:
    for line in f:
        question=json.loads(line)
        q_dict={}
        temp_str=question['question']['stem'].encode('utf-8').strip()
        q_dict['Q']=re.sub("[^\w]", " ",  temp_str).split()
        q_dict['Q'] = parse_func(q_dict['Q'])
        q_dict['positive'] = "Yes"
        for neg_word in not_words:
            if neg_word in q_dict['Q']:
                q_dict['positive'] = "No"
                break
        for opt in question['question']['choices']:
            temp_str=opt['text'].encode('utf-8').strip()
            q_dict[opt['label'].encode('utf-8').strip()]=re.sub("[^\w]", " ",  temp_str).split()
            q_dict[opt['label'].encode('utf-8').strip()]=parse_func(q_dict[opt['label'].encode('utf-8').strip()])
        data.append(q_dict)
```
